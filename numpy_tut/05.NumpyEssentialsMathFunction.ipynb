```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sin 곡선 구하기. 
# x 축은 0 ~ 10까지의 범위를 0.05 간격으로 데이터를 생성한다. 
x = np.arange(0, 10, 0.05)
# y 축은 x축에서 sin 결과를 출력한다. (sin특성상 1에서 -1 범위의 값이 출력된다.)
y = np.sin(x)

plt.plot(x, y)
```

```python
# cos 곡선 구하기. 
# x 축은 0 ~ 10까지의 범위를 0.05 간격으로 데이터를 생성한다. 
x = np.arange(0, 10, 0.05)
# y 축은 x축에서 cos 결과를 출력한다. (cos특성상 1에서 -1 범위의 값이 출력된다. sin과 반대로 움직인다.)
y = np.cos(x)

plt.plot(x, y)
```

```python
# sin, cos 그래프 2개 그래프 함께 그리기 
x = np.arange(0, 10, 0.05)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_cos)
plt.plot(x, y_sin)
```

```python
# tan 그래프 그리기 
# sin, cos 그래프 2개 그래프 함께 그리기 
x = np.arange(-1.5, 1.5, 0.01)

y_tan = np.tan(x)

plt.plot(x, y_tan)
```

```python
# log 그래프 
x = np.linspace(0, 51, 50)
y = np.log(x)

plt.plot(x, y)
```

```python
# 선형 그래프 
x = np.linspace(0, 51, 50)
y = x

plt.plot(x, y)
```
